### 1️⃣ Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.feature_selection import SelectKBest, chi2
```

### 2️⃣ Load Dataset

```python
df = pd.read_csv(r'D:\juypterNoteBook\Machine learing\collage\adult.csv')
df.head(10)
```

```python
df.info()
df.shape
```

### 3️⃣ Data Cleaning

#### 3.1 Handle Missing Values (?)

```python
import numpy as np
df.replace('?', np.nan, inplace=True)
```

```python
df.isnull().sum()
```

#### 3.2 Fill Missing Values

```python
df['workclass'].fillna(df['workclass'].mode()[0], inplace=True)
df['occupation'].fillna(df['occupation'].mode()[0], inplace=True)
df['native-country'].fillna(df['native-country'].mode()[0], inplace=True)
```

#### 3.3 Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

### 4️⃣ Data Transformation

4.1 Encode Target Variable

```python
df['income'] = df['income'].map({'<=50K':0, '>50K':1})
```

4.2 Encode Categorical Features

```python
le = LabelEncoder()

for col in df.select_dtypes(include='object').columns:
    df[col] = le.fit_transform(df[col])
```

### 5️⃣ Data Reduction

5.1 Remove Redundant Features
education and educational-num represent same information

```python
df.drop('education', axis=1, inplace=True)
```

5.2 Feature Selection (Chi-Square)

```python
X = df.drop('income', axis=1)
y = df['income']

selector = SelectKBest(score_func=chi2, k=8)
X_new = selector.fit_transform(X, y)

selected_features = X.columns[selector.get_support()]
print("Selected Features:", selected_features)
```

### 6️⃣ Data Discretization

Convert continuous data into categories.

6.1 Age Binning

```python
df['age_group'] = pd.cut(df['age'],
                         bins=[0,25,45,65,100],
                         labels=['Young','Adult','Senior','Old'])
```

6.2 Hours-per-week Binning

```python
df['work_hours_group'] = pd.cut(df['hours-per-week'],
                                bins=[0,25,40,60,100],
                                labels=['Part-time','Full-time','Overtime','Heavy'])
```

### 7️⃣ Data Normalization

7.1 Min-Max Scaling (0–1)

```python
scaler = MinMaxScaler()
num_cols = ['age','fnlwgt','capital-gain','capital-loss','hours-per-week']
df[num_cols] = scaler.fit_transform(df[num_cols])
```

OR Standardization (Z-score)

```python
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

### 8️⃣ Data Integration

```python
df['net_capital'] = df['capital-gain'] - df['capital-loss']
df.drop(['capital-gain','capital-loss'], axis=1, inplace=True)
```
